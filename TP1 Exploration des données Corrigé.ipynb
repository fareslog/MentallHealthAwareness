### 3.  Mesure des caractéristiques statistiques

`````python

# Calcul de la Moyenne
def calculer_moyenne(df, colonne):
    return df[colonne].mean()

#  Calcul de la Médiane
def calculer_mediane(df, colonne):
    return df[colonne].median()

#  Calcul des Quartiles
def calculer_quartiles(df, colonne):
    Q1 = df[colonne].quantile(0.25)
    Q2 = df[colonne].median()
    Q3 = df[colonne].quantile(0.75)
    return Q1, Q2, Q3

#  Calcul de la Variance et de l'Écart-type
def calculer_variance_et_ecart_type(df, colonne):
    variance = df[colonne].var()
    ecart_type = df[colonne].std()
    return variance, ecart_type

#  Calcul de l'Étendue Interquartile (IQR)
def calculer_etendue_interquartile(Q1, Q3):
    return Q3 - Q1

# Liste des colonnes à analyser
colonnes = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Exécuter les calculs et afficher les résultats pour chaque caractéristique
for colonne in colonnes:
    print(f"\nCaractéristique: {colonne}")
    
    # Moyenne
    moyenne = calculer_moyenne(df, colonne)
    print(f"Moyenne : {moyenne:.2f}")
    
    # Médiane
    mediane = calculer_mediane(df, colonne)
    print(f"Médiane : {mediane:.2f}")
    
    # Quartiles
    Q1, Q2, Q3 = calculer_quartiles(df, colonne)
    print(f"Quartile 1 (Q1) : {Q1:.2f}")
    print(f"Quartile 3 (Q3) : {Q3:.2f}")
    
    # Variance et Ecart-type
    variance, ecart_type = calculer_variance_et_ecart_type(df, colonne)
    print(f"Variance : {variance:.2f}")
    print(f"Écart-type : {ecart_type:.2f}")
    
    # Étendue Interquartile
    etendue_interquartile = calculer_etendue_interquartile(Q1, Q3)
    print(f"Étendue interquartile (IQR) : {etendue_interquartile:.2f}")
`````

### 4. Détection des valeurs aberrantes

`````python
######1ére Méthode: fonction pour détecter les position des valeurs abérentes et leur nombre

# Détection des outliers et remplacement par la moyenne
for column in df.columns[:-1]: # Exclure la colonne 'species'
    # Calcul des quartiles et de l'IQR
    print(column)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Définir les limites des outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Détecter les positions des outliers 
    outlier_positions = ((df[column] < lower_bound) | (df[column] > upper_bound))
    print(outlier_positions.sum())
`````

`````python
##### 2ème méthode: Créer des boxplots pour chaque colonne 
for column in df.columns[:-1]: # Exclure la colonne 'species' 
    plt.figure(figsize=(8, 4)) 
    sns.boxplot(x=df[column]) 
    plt.title(f'Boxplot de {column}')
    plt.xlabel(column)
`````
